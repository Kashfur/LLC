**Concating dataframes having monthly data to create one dataframe**

```python
df = pd.DataFrame()
df_bymonth = [WP_df, UN_df, M_df, med_df,infl_df, CONSTR_df, pop_df,con_df,gdp_df,hp_df]
for df1 in df_bymonth:
    df1["DATE"] = pd.to_datetime(df1["DATE"])
    df1 = df1.set_index("DATE")
    df = pd.concat([df,df1], axis = 1)
print(df.shape)
df.head()
```

# **Data Cleaning**

```python
df.shape
```

```python
df.info()
```

```python
#checking null values
df.isnull().sum()
```

```python
# Filling missing values in GDP and House price index column using linear interpolation
df["Gdp"] = df["Gdp"].interpolate()
df['House Price Index']=df['House Price Index'].interpolate()
```

```python
df.describe()
```

```python
df['CONSUMER PRICE INDEX'].fillna(np.mean,inplace=True)
df['CONSTRUCTION SPENDING'].fillna(np.mean,inplace=True)
df['Monthly new house supply'].fillna(np.mean,inplace=True)
df['Working_population'].fillna(np.mean,inplace=True)
df['unemployement rate'].fillna(np.mean,inplace=True)
```

```python
df['Inflation']=df['Inflation'].interpolate()
```

```python
df['Median Household income']=df['Median Household income'].interpolate()
```

```python
df.isnull().sum()
```

# **Factors that could influence Home Prices**

```python
#checking correlation
import matplotlib.pyplot as plt
import seaborn as sns
# Calculate the correlation matrix
correlation_matrix = df.corr()

#Display the correlation matrix
print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 8))

# Plotting the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Variables')
plt.show()
```

In this heatmap:

The values closer to 1 indicate a strong positive correlation.
Values closer to -1 indicate a strong negative correlation.
Values close to 0 suggest little to no correlation.
